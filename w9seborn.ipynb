```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for 
# different events and regions.
import seaborn as sns
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.relplot(x=fmri.timepoint,y=fmri.signal)
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and 
# y = 'fare'.
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot a box plot for 'age' with 'pclass' as the x-axis on the first subplot
sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0])
axes[0].set_title("Box plot of Age by Pclass")

# Plot a box plot for 'fare' with 'pclass' as the x-axis on the second subplot
sns.boxplot(x='pclass', y='fare', data=titanic, ax=axes[1])
axes[1].set_title("Box plot of Fare by Pclass")

# Display the plots
plt.tight_layout()
plt.show()

```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
flights=sns.load_dataset('flights')
flights
```

```python

import matplotlib.pyplot as plt

# Pivot the dataset to have years as rows and months as columns
flights_pivot = flights.pivot("year", "month", "passengers")

# Create a heatmap
plt.figure(figsize=(12, 2))
sns.heatmap(flights_pivot, annot=True, cmap="YlGnBu", fmt="d", linewidths=.5)
plt.title("Monthly Passenger Counts (in thousands) Over Years")

plt.show()
```
