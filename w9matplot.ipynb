```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# in the scatter plot according to the coordinate it plot the point onn the graph 
```

```python
plt.scatter(x,y)
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.scatter(x,y)
```
